```python
import os, unidecode, datetime, requests, json
from dotenv import load_dotenv
today = datetime.datetime.today()


load_dotenv()

league_name  = "La Liga"
calendar_url = "https://www.laliga.com/en-FR/laliga-easports/results/2023-24/gameweek-"
tags         = {
    "match_tag":  "table > tbody > tr:nth-child(3n+1)",
    "date_tag":   "td:nth-child(2)",
    "hour_tag":   "td:nth-child(3)",
    "teams_tag":  "td:nth-child(5) a",
}
NOTION_ENDPOINT = "https://api.notion.com/v1/"



NOTION_API_KEY = os.getenv("NOTION_API_KEY")
headers = {
    "accept": "application/json",
    "Notion-Version": "2022-06-28",
    "content-type": "application/json",
    "Authorization": "Bearer " + NOTION_API_KEY 
}

MATCHES_DB_ID  = os.getenv("MATCHES_DB_ID")
league_id = os.getenv(unidecode.unidecode(league_name.replace(" ", "").lower()))

data = {
    # "Home Team":"d759a042c2334e6da83bea9fb1c500fc",
    # "Home Team":"d759a042c2334e6da83bea9fb1c500fc",
    "filter":{
        "and": [
            {
                "property": "League",
                "relation": {"contains": league_id}
            },

            {
                "property": "Played?",
                "checkbox": {"equals": False}
            },
            {
                "property": "Date",
                "date": {"before": today.strftime("%Y-%m-%d %H:%M")}
            }
        ]
    },

    "sorts": [
        {
            "property": "Date",
            "direction": "ascending"
        }
    ]
}
    
fixtures = requests.post(
    url=NOTION_ENDPOINT+"databases/"+MATCHES_DB_ID+"/query",
    headers=headers,
    json=data
    ).json()

for item in fixtures["results"]:
    data = item["properties"]
    match_name = data.get("Name").get("title")[0].get("text").get("content")
    match_id = item.get("id")
    home_team = data.get("Home Team").get("relation")[0].get("id")
    print(home_team)

    # update = requests.patch(
    #     url = NOTION_ENDPOINT+"pages/"+match_id,
    #     headers=headers,
    #     json={
    #         "properties": {
    #             "Played?": {"checkbox": True}
    #         }
    #     }
    # )
```
